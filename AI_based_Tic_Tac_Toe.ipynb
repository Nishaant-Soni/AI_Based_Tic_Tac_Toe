```python
import tkinter as tk
from tkinter import messagebox
import random

class TicTacToe:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Tic Tac Toe")
        self.board = [' ' for _ in range(9)]
        self.buttons = []
        self.create_buttons()
        self.choose_player()
        self.window.mainloop()

    def create_buttons(self):
        for i in range(9):
            button = tk.Button(self.window, text=' ', font='normal 20 bold', height=3, width=6,
                               command=lambda i=i: self.on_button_click(i))
            button.grid(row=i//3, column=i%3)
            self.buttons.append(button)

    def choose_player(self):
        self.player_choice = tk.StringVar()
        self.ai_choice = tk.StringVar()
        self.player_choice.set('X')
        self.ai_choice.set('O')
        
        player_label = tk.Label(self.window, text="Choose your player:")
        player_label.grid(row=3, column=0, columnspan=3)
        
        x_button = tk.Radiobutton(self.window, text="X", variable=self.player_choice, value='X', command=self.set_player)
        x_button.grid(row=4, column=0)
        
        o_button = tk.Radiobutton(self.window, text="O", variable=self.player_choice, value='O', command=self.set_player)
        o_button.grid(row=4, column=1)
        
        start_button = tk.Button(self.window, text="Start Game", command=self.start_game)
        start_button.grid(row=4, column=2)

    def set_player(self):
        self.ai_choice.set('O' if self.player_choice.get() == 'X' else 'X')

    def start_game(self):
        if self.player_choice.get() == 'O':
            self.current_player = 'O'
            self.ai_move()
        else:
            self.current_player = 'X'

    def on_button_click(self, index):
        if self.board[index] == ' ' and self.current_player == self.player_choice.get():
            self.board[index] = self.player_choice.get()
            self.buttons[index].config(text=self.player_choice.get())
            if self.check_winner(self.player_choice.get()):
                messagebox.showinfo("Tic Tac Toe", f"Player {self.player_choice.get()} wins!")
                self.reset_game()
            elif ' ' not in self.board:
                messagebox.showinfo("Tic Tac Toe", "It's a tie!")
                self.reset_game()
            else:
                self.current_player = self.ai_choice.get()
                self.ai_move()

    def ai_move(self):
        best_score = float('-inf') if self.ai_choice.get() == 'O' else float('inf')
        best_move = None
        for i in range(9):
            if self.board[i] == ' ':
                self.board[i] = self.ai_choice.get()
                score = self.minimax(self.board, 0, self.ai_choice.get() == 'X', float('-inf'), float('inf'))
                self.board[i] = ' '
                if (self.ai_choice.get() == 'O' and score > best_score) or (self.ai_choice.get() == 'X' and score < best_score):
                    best_score = score
                    best_move = i
        self.board[best_move] = self.ai_choice.get()
        self.buttons[best_move].config(text=self.ai_choice.get())
        if self.check_winner(self.ai_choice.get()):
            messagebox.showinfo("Tic Tac Toe", f"Player {self.ai_choice.get()} wins!")
            self.reset_game()
        elif ' ' not in self.board:
            messagebox.showinfo("Tic Tac Toe", "It's a tie!")
            self.reset_game()
        else:
            self.current_player = self.player_choice.get()

    def minimax(self, board, depth, is_maximizing, alpha, beta):
        if self.check_winner(self.ai_choice.get()):
            return 1 if self.ai_choice.get() == 'O' else -1
        elif self.check_winner(self.player_choice.get()):
            return -1 if self.ai_choice.get() == 'O' else 1
        elif ' ' not in board:
            return 0

        if is_maximizing:
            max_eval = float('-inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = self.ai_choice.get()
                    eval = self.minimax(board, depth + 1, False, alpha, beta)
                    board[i] = ' '
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
            return max_eval
        else:
            min_eval = float('inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = self.player_choice.get()
                    eval = self.minimax(board, depth + 1, True, alpha, beta)
                    board[i] = ' '
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
            return min_eval

    def check_winner(self, player):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for condition in win_conditions:
            if self.board[condition[0]] == self.board[condition[1]] == self.board[condition[2]] == player:
                return True
        return False

    def reset_game(self):
        self.board = [' ' for _ in range(9)]
        for button in self.buttons:
            button.config(text=' ')
        self.choose_player()

if __name__ == "__main__":
    TicTacToe()
```

```python
import abc
import time
import random
import numpy as np

class AIStrategy(abc.ABC):
    @abc.abstractmethod
    def get_move(self, board, player):
        pass

class MinimaxStrategy(AIStrategy):
    def get_move(self, board, player):
        best_score = float('-inf') if player == 'O' else float('inf')
        best_move = None
        for i in range(9):
            if board[i] == ' ':
                board[i] = player
                score = self.minimax(board, 0, player == 'X')
                board[i] = ' '
                if (player == 'O' and score > best_score) or (player == 'X' and score < best_score):
                    best_score = score
                    best_move = i
        return best_move

    def minimax(self, board, depth, is_maximizing):
        if self.check_winner(board, 'O'):
            return 1
        elif self.check_winner(board, 'X'):
            return -1
        elif ' ' not in board:
            return 0

        if is_maximizing:
            max_eval = float('-inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = 'O'
                    eval = self.minimax(board, depth + 1, False)
                    board[i] = ' '
                    max_eval = max(max_eval, eval)
            return max_eval
        else:
            min_eval = float('inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = 'X'
                    eval = self.minimax(board, depth + 1, True)
                    board[i] = ' '
                    min_eval = min(min_eval, eval)
            return min_eval

    def check_winner(self, board, player):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for condition in win_conditions:
            if board[condition[0]] == board[condition[1]] == board[condition[2]] == player:
                return True
        return False

class AlphaBetaStrategy(AIStrategy):
    def get_move(self, board, player):
        best_score = float('-inf') if player == 'O' else float('inf')
        best_move = None
        for i in range(9):
            if board[i] == ' ':
                board[i] = player
                score = self.alphabeta(board, 0, player == 'X', float('-inf'), float('inf'))
                board[i] = ' '
                if (player == 'O' and score > best_score) or (player == 'X' and score < best_score):
                    best_score = score
                    best_move = i
        return best_move

    def alphabeta(self, board, depth, is_maximizing, alpha, beta):
        if self.check_winner(board, 'O'):
            return 1
        elif self.check_winner(board, 'X'):
            return -1
        elif ' ' not in board:
            return 0

        if is_maximizing:
            max_eval = float('-inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = 'O'
                    eval = self.alphabeta(board, depth + 1, False, alpha, beta)
                    board[i] = ' '
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
            return max_eval
        else:
            min_eval = float('inf')
            for i in range(9):
                if board[i] == ' ':
                    board[i] = 'X'
                    eval = self.alphabeta(board, depth + 1, True, alpha, beta)
                    board[i] = ' '
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
            return min_eval

    def check_winner(self, board, player):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for condition in win_conditions:
            if board[condition[0]] == board[condition[1]] == board[condition[2]] == player:
                return True
        return False

class QLearningStrategy(AIStrategy):
    def __init__(self):
        self.q_table = {}
        self.learning_rate = 0.1
        self.discount_factor = 0.9
        self.exploration_rate = 1.0
        self.exploration_decay = 0.99

    def get_move(self, board, player):
        state = tuple(board)
        if state not in self.q_table:
            self.q_table[state] = [0] * 9

        if random.uniform(0, 1) < self.exploration_rate:
            move = random.choice([i for i in range(9) if board[i] == ' '])
        else:
            move = np.argmax(self.q_table[state])

        self.exploration_rate *= self.exploration_decay
        return move

    def update_q_table(self, state, action, reward, next_state):
        if state not in self.q_table:
            self.q_table[state] = [0] * 9
        if next_state not in self.q_table:
            self.q_table[next_state] = [0] * 9

        best_next_action = np.argmax(self.q_table[next_state])
        self.q_table[state][action] = self.q_table[state][action] + self.learning_rate * (
            reward + self.discount_factor * self.q_table[next_state][best_next_action] - self.q_table[state][action]
        )

# Integrate these strategies into the Tic Tac Toe game
class TicTacToe:
    def __init__(self, ai_strategy):
        self.window = tk.Tk()
        self.window.title("Tic Tac Toe")
        self.board = [' ' for _ in range(9)]
        self.current_player = 'X'
        self.buttons = []
        self.ai_strategy = ai_strategy
        self.create_buttons()
        self.start_game()
        self.window.mainloop()

    def create_buttons(self):
        self.game_frame = tk.Frame(self.window)
        self.game_frame.grid(row=0, column=0, columnspan=3)
        self.canvas = tk.Canvas(self.game_frame, width=300, height=300, bg='white')
        self.canvas.grid(row=0, column=0)
        self.canvas.create_line(100, 0, 100, 300, width=2, fill='white')
        self.canvas.create_line(200, 0, 200, 300, width=2, fill='white')
        self.canvas.create_line(0, 100, 300, 100, width=2, fill='white')
        self.canvas.create_line(0, 200, 300, 200, width=2, fill='white')
        for i in range(9):
            button = tk.Button(self.game_frame, text=' ', font='normal 20 bold', height=1, width=3, bg='white', relief='flat', command=lambda i=i: self.on_button_click(i))
            x = (i % 3) * 100
            y = (i // 3) * 100
            button.place(x=x, y=y, width=100, height=100)
            self.buttons.append(button)

    def start_game(self):
        if random.choice(['X', 'O']) == 'O':
            self.current_player = 'O'
            self.ai_move()

    def on_button_click(self, index):
        if self.board[index] == ' ' and self.current_player == 'X':
            self.board[index] = 'X'
            self.buttons[index].config(text='X')
            if self.check_winner('X'):
                messagebox.showinfo("Tic Tac Toe", "Player X wins!")
                self.reset_game()
            elif ' ' not in self.board:
                messagebox.showinfo("Tic Tac Toe", "It's a tie!")
                self.reset_game()
            else:
                self.current_player = 'O'
                self.ai_move()

    def ai_move(self):
        move = self.ai_strategy.get_move(self.board, 'O')
        self.board[move] = 'O'
        self.buttons[move].config(text='O')
        if self.check_winner('O'):
            messagebox.showinfo("Tic Tac Toe", "Player O wins!")
            self.reset_game()
        elif ' ' not in self.board:
            messagebox.showinfo("Tic Tac Toe", "It's a tie!")
            self.reset_game()
        else:
            self.current_player = 'X'

    def check_winner(self, player):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for condition in win_conditions:
            if self.board[condition[0]] == self.board[condition[1]] == self.board[condition[2]] == player:
                return True
        return False

    def reset_game(self):
        self.board = [' ' for _ in range(9)]
        for button in self.buttons:
            button.config(text=' ')
        self.current_player = 'X'
        self.start_game()

if __name__ == "__main__":
    strategy = QLearningStrategy()  # Change this to AlphaBetaStrategy(), MCTSStrategy(), or QLearningStrategy() to test different strategies
    TicTacToe(strategy)
```
